```
import jax.numpy as jnp

from smooth_binary_node import SmoothBinaryNode


node_left = SmoothBinaryNode()
node_right = SmoothBinaryNode()
root = SmoothBinaryNode(left_node=node_left, right_node=node_right)

left_params = {'weights': jnp.array([0.0, 10.0, 0.0]),
               'leaves': jnp.array([1.0, 2.0]),
               'bias': 25.0}

right_params = {'weights': jnp.array([0.0, 10.0, 0.0]),
                'leaves': jnp.array([3.0, 4.0]),
                'bias': 75.0}

params = {'weights': jnp.array([0.0, 10.0, 0.0]),
          'leaves': None,
          'bias': 50.0,
          'left_params': left_params,
          'right_params': right_params}

features = jnp.array([[1, 80, 7]])
pred = root.predict(params, features)
# 80 is greater than 50
# and greater than 75
# we expect a 4.0
print('two levels', pred)
#> 4.0

features = jnp.array([[1, 70, 7]])
pred = root.predict(params, features)
# 80 is greater than 50
# and lesser than 75
# we expect a 3.0
print('two levels', pred)
#> 3.0

features = jnp.array([[1, 45, 7]])
pred = root.predict(params, features)
# 24 is lesser than 50
# and greater than 25
# we expect a 2.0
print('two levels', pred)
#> 2.0

features = jnp.array([[1, 20, 7]])
pred = root.predict(params, features)
# 24 is lesser than 50
# and lesser than 25
# we expect a 1.0
print('two levels', pred)
#> 1.0
```
